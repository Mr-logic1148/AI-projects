```python
import random
import time
#Credits : https://www.educative.io/answers/how-to-play-an-audio-file-in-pygame
from pygame import mixer
import pygame
mixer.init()
movement_sound = pygame.mixer.Sound("movement.mp3")
#victory_sound = pygame.mixer.Sound("victory_effect.mp3")
#fail_sound = pygame.mixer.Sound("fail.mp3")
error_message = pygame.mixer.Sound("error_message.mp3")
```

```python
def board():
    letters = ['x','o','x','o','x','o','o','o','x']
    print("""\n
     {} | {} | {} 
    ---+---+---
     {} | {} | {} 
    ---+---+---
     {} | {} | {} 
    """.format(letters[0],letters[1],letters[2],letters[3],letters[4],letters[5],letters[6],letters[7],letters[8]))
```

```python
def win_conditions(total_move):
    if total_move[0] == total_move[1] == total_move[2]:
        return total_move[0]
    elif total_move[0] == total_move[3] == total_move[6]:
        return total_move[0]
    elif total_move[0] == total_move[4] == total_move[8]:
        return total_move[0]
    elif total_move[1] == total_move[4] == total_move[7]:
        return total_move[1]
    elif total_move[2] == total_move[5] == total_move[8]:
        return total_move[2]
    elif total_move[3] == total_move[4] == total_move[5]:
        return total_move[3]
    elif total_move[6] == total_move[7] == total_move[8]:
        return total_move[6]
    elif total_move[2] == total_move[4] == total_move[6]:
        return total_move[2]
    else:
        return None
```

```python
def play(character_choice):
    board = """\n
     {} | {} | {} 
    ---+---+---
     {} | {} | {} 
    ---+---+---
     {} | {} | {} 
    """
    counter = 0
    moves_list = [0,1,2,3,4,5,6,7,8]
    total_move = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    human_player = character_choice
    if human_player == "x":
        bot_player = "o"
    else:
        bot_player = "x"
    print("!THIS IS THE LOCATION CHART TO HELP YOU ENTER YOUR MOVES!")
    locations()
    while counter < 5:
        flag = 0
        while flag == 0:
            print(flag)
            move = input("\nEnter location of your move:")
            if move.isdigit():
                m = int(move)
                print(total_move[m - 1])
                if total_move[m - 1] == ' ':
                    pygame.mixer.Sound.play(movement_sound)
                    flag = 1
                else:
                    pygame.mixer.Sound.play(error_message)
                    time.sleep(2)
                    print("\nLOCATION OCCUPIED!!")
            else:
                pygame.mixer.Sound.play(error_message)
                time.sleep(2)
                print("\nINVALID LOCATION PLEASE TRY AGAIN!!!")
        print(flag)
        h_move = m - 1
        total_move[h_move] = human_player
        moves_list.remove(h_move)
        print(board.format(total_move[0],total_move[1],total_move[2],total_move[3],total_move[4],total_move[5],total_move[6],total_move[7],total_move[8]))
        counter = counter + 1
        if counter > 2:
            print("\nwin 1")
            winner = win_conditions(total_move)
            if winner == human_player:
                print("!!! BRAVO YOU WON THE GAME !!!")
                break
            elif winner == bot_player:
                print("!!! SORRY AI WON THE GAME !!!")
                break
        print("AI IS PLAYING ITS TURN...")
        pygame.mixer.Sound.play(movement_sound)
        time.sleep(3)
        bot_move = random.choice(moves_list)
        total_move[bot_move] = bot_player
        moves_list.remove(bot_move)
        print(board.format(total_move[0],total_move[1],total_move[2],total_move[3],total_move[4],total_move[5],total_move[6],total_move[7],total_move[8]))
        if counter > 2:
            print("\nwin 2")
            winner = win_conditions(total_move)
            if winner == human_player:
                print("!!! BRAVO YOU WON THE GAME !!!")
                break
            elif winner == bot_player:
                print("!!! SORRY AI WON THE GAME !!!")
                break
```
