```python
def create_table(size):
    """THIS FUNCTION IS USED TO CREATE THE GRID OF DIFFERENT SIZE
    ______________________________________________________________
    INPUT : SIZE
    OUTPUT: TABLE"""
    for i in range(size):
        top = "+---"*size + "+\n"
        mid = "| {} "*size + "|"
    t_m = top+mid+"\n"
    base = "+---"*size + "+\n"
    table = t_m * size + base
    return table
```
